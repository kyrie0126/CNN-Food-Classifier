##### Food Choices: sashimi and pho

```python
# decide which two foods to classify
labelA = 'sashimi'
labelB = 'pho'
```

##### **Some Data Preparation**
Next, we're gonna do some preparation to amke the data handling easier. We're gonna do two things:
1. remove all images that don't have labelA or labelB
2. rename images with labelA or labelB to have that label in their filename

```python
# Loop through all images downloaded
for img in get_image_files(foodPath):
  # rename images so that the labelA or labelB is in the file name
  if labelA in str(img):
    img.rename(f"{img.parent}/{labelA}_{img.name}")
  elif labelB in str(img):
    img.rename(f"{img.parent}/{labelB}_{img.name}")
  # remove images that aren't labelA or labelB
  else:
    os.remove(img)
```

```python
len(get_image_files(foodPath))
# we can see that the dataset is reduced to 1K of each label
```

## Train Model

```python
def GetLabel(filename):
  return filename.split('_')[0]

# test
GetLabel('pho_846615.jpg')
```

```python
dls = ImageDataLoaders.from_name_func(
    foodPath,
    get_image_files(foodPath),
    valid_pct=0.2,
    seed=420,
    label_func=GetLabel,
    item_tfms=Resize(224)
)

dls.valid.show_batch()
```

```python
learn = cnn_learner(
    dls,
    resnet34,
    metrics=error_rate,
    pretrained=True
)
learn.fine_tune(epochs=10)
```

```python
from google.colab import files
uploader = files.upload()

for img in uploader.items():
  uploadedImg = img[0]

img = PILImage.create(uploadedImg)
img.show()

label,_,probs = learn.predict(img)

print(f"This is a {label}.")
print(f"{labelA} {probs[1].item():.6f}")
print(f"{labelB} {probs[0].item():.6f}")
```

## Confusion Matrix of Model Performance

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
interp.plot_top_losses(7)
```
